```python
import io, urllib.request, pandas as pd
from datetime import datetime, timedelta
from pytz import timezone
from utils import to_ts_range

start = datetime(year=2026, month=3, day=13, hour=16, minute=0)
window = timedelta(minutes=15)
s, e = to_ts_range(timezone("America/New_York").localize(start), window)

url = f"https://csi-fs-pi-data-server.ngrok.io/export/0/all/FsLoxGn2Transducers/records?&startTs={s}&endTs={e}"
req = urllib.request.Request(url, headers={"ngrok-skip-browser-warning": "true"})
with urllib.request.urlopen(req) as r:
    raw = r.read()

df = pd.read_csv(io.BytesIO(raw))
print(df.columns.tolist())
print(df.head(2))
```
